```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from scipy.stats import gamma
import seaborn as sns
```

```python
xs = np.array([5, 10, 15, 20, 25])
n = 30

alphas = xs + 1
betas = n - xs + 1

ps = np.linspace(0, 1, 10**5)

colors = sns.color_palette("Set2", 5)

for x, a, b, color in zip(xs, alphas, betas, colors):
    plt.plot(ps, beta.pdf(ps, a, b), label=f'{x}', color=color)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, title='x', title_fontsize='12', fontsize='10', loc='upper right', frameon=True)

plt.plot(ps, np.ones(len(ps)), color='red', linestyle='--', linewidth=2)

plt.xlabel('$\pi$', fontsize=14, style='italic', labelpad = 10)
plt.ylabel('Densit di probabilit', fontsize=14, style='italic', labelpad = 10)

plt.xlim(0, 1.15)
plt.show()
```

```python
xs = np.array([5, 10, 15, 20, 25])
n = 30
alpha_p = 7
beta_p  = 4

alphas = xs + alpha_p
betas = n - xs + beta_p

ps = np.linspace(0, 1, 10**5)

colors = sns.color_palette("Set2", 5)

for x, a, b, color in zip(xs, alphas, betas, colors):
    plt.plot(ps, beta.pdf(ps, a, b), label=f'{x}', color=color)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, title='x', title_fontsize='12', fontsize='10', loc='upper right', frameon=True)

plt.plot(ps,beta.pdf(ps, alpha_p, beta_p), color='red', linestyle='--', linewidth=2)

plt.xlabel('$\pi$', fontsize=14, style='italic', labelpad = 10)
plt.ylabel('Densit di probabilit', fontsize=14, style='italic', labelpad = 10)

plt.xlim(0, 1.15)
plt.show()
```

```python
xs = np.array([32, 44, 35, 29, 26, 39])

k  = np.sum(xs) + 1
mu = len(xs)

lambdas = ps = np.linspace(0, 50, 10**4)
pdf_values = gamma(a=k, scale=1/mu).pdf(lambdas)
prior_values = np.array([1/max(lambdas) for _ in range(len(lambdas))])

plt.plot(lambdas, pdf_values)
plt.plot(lambdas, prior_values, color='red', linestyle='--', linewidth=2)
plt.show()
```
